```python
# Initialize a single figure
plt.figure(figsize=(12, 8))

# Loop through each clone and plot its data
for c in range(100):
    try:
        # Attempt to load the data
        all_all_frame_energies = np.load(f'../EE_analysis/RUN2_CLONE{c}_feb.npy')
        print(f"Plot ../EE_analysis/RUN2_CLONE{c}_feb.npy")
    except FileNotFoundError:
        # Skip this iteration if the file is not found
        print(f"File not found: ../EE_analysis/RUN2_CLONE{c}_feb.npy")
        continue

    
    # Plot the data for this clone
    time_ns = np.array(range(len(all_all_frame_energies))) / 100.0  # Convert time to ns
    free_energies = all_all_frame_energies[:, -1]
    
    # Scatter plot for the free energy values
    plt.scatter(time_ns, free_energies, s=6, alpha=0.6, label=f'Clone {c}' if c < 100 else '')  # Show label for first 10 clones

# Add labels, title, and legend
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('Free Energies', fontsize=12)
plt.title('Free Energy Differences for All Clones', fontsize=14)
plt.axhline(y=np.average(free_energies), color='red', linestyle='--', linewidth=1.2, label='Average Energy')
plt.legend(loc='upper right', ncol=2, fontsize=8, frameon=False)  # Compact legend
plt.tight_layout()

# Show the plot
plt.show()
```
