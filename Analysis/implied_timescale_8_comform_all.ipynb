```python
import numpy as np
from matplotlib import pyplot as plt
import pickle
import pandas as pd
from tqdm import tqdm 
import bottleneck as bn
```

## Load Important Functoin

```python
class Data:
    def __init__(self, array_list):
        self.array_list = array_list

    def save(self, filename):
        with open(filename, 'wb') as f:
            pickle.dump(self.array_list, f)

    @classmethod
    def load(cls, filename):
        with open(filename, 'rb') as f:
            array_list = pickle.load(f)
        return cls(array_list)
    
```

## Plain MD

```python
plain_md_path = '../Plain_MD/analysis'


omega_loaded = Data.load(f"{plain_md_path}/autocoorel_data/stationary_distributions_1800ns_5ns_clustersize100.pkl").array_list
pi_all_means = omega_loaded[0]
pi_all_uppers = omega_loaded[1]
pi_all_lowers = omega_loaded[2]
pi_all_stds = omega_loaded[3]
t2_all_means = omega_loaded[4]
t2_all_uppers = omega_loaded[5]
t2_all_lowers = omega_loaded[6]
t2_all_stds = omega_loaded[7]
lagtimes_ns = omega_loaded[8]

#pi_all_means
plain_md_lagtimes_ns = lagtimes_ns
plain_md_t2_all_means = t2_all_means
plain_md_t2_all_uppers = t2_all_uppers
plain_md_t2_all_lowers = t2_all_lowers
plain_md_t2_all_stds = t2_all_stds
```

```python
dt_in_ns = 0.01

plt.figure()
plt.plot(plain_md_lagtimes_ns, plain_md_t2_all_means*dt_in_ns, 'ko-', alpha=1)
plt.fill_between(plain_md_lagtimes_ns, plain_md_t2_all_uppers*dt_in_ns, plain_md_t2_all_lowers*dt_in_ns, alpha=0.3)
plt.xlabel('lagtime (ns)')
plt.ylabel('implied timescale (ns)')

plt.yscale('log')
plt.ylim(5, 100000)
plt.title(f'Implied timescales for Plain MD')
plt.show()
```

```python
dt_in_ns = 0.01

plt.figure()
plt.plot(lagtimes_ns, plain_md_t2_all_means*dt_in_ns, 'ko-')
plt.fill_between(lagtimes_ns, plain_md_t2_all_uppers*dt_in_ns, plain_md_t2_all_lowers*dt_in_ns, alpha=0.3)
plt.xlabel('lagtime (ns)')
plt.ylabel('implied timescale (ns)')

plt.yscale('log')
plt.xscale('log')
plt.ylim(5, 100000)
plt.title(f'Implied timescales for Plain MD')
plt.show()
```

## Combined Plot

```python
import matplotlib.cm as cm

# Get 4 colors evenly spaced from the Viridis colormap
viridis = cm.get_cmap('viridis')
colors = [viridis(i) for i in np.linspace(0, 1, 4)]

plt.figure(figsize=(8, 6))

# Plain MD
dt_plain = 0.01
plt.plot(plain_md_lagtimes_ns, plain_md_t2_all_means*dt_plain, '-', marker='o', color=colors[0], label='Plain MD', alpha=0.2)
plt.fill_between(plain_md_lagtimes_ns, plain_md_t2_all_uppers*dt_plain, plain_md_t2_all_lowers*dt_plain, color=colors[0], alpha=0.2)

# EE
dt_ee = 0.01
plt.plot(ee_lagtimes_ns, ee_t2_all_means*dt_ee, '-', marker='s', color=colors[1], label='EE', alpha=0.2)
plt.fill_between(ee_lagtimes_ns, ee_t2_all_uppers*dt_ee, ee_t2_all_lowers*dt_ee, color=colors[1], alpha=0.2)

# HREX
dt_hrex = 0.01
plt.plot(hrex_lagtimes_ns, hrex_t2_all_means*dt_hrex, '-', marker='^', color=colors[2], label='HREX', alpha=0.2)
plt.fill_between(hrex_lagtimes_ns, hrex_t2_all_uppers*dt_hrex, hrex_t2_all_lowers*dt_hrex, color=colors[2], alpha=0.2)

# REDs
dt_reds = 0.004
plt.plot(reds_lagtimes_ns, reds_t2_all_means*dt_reds, '-', marker='D', color=colors[3], label='REDs', alpha=0.2)
plt.fill_between(reds_lagtimes_ns, reds_t2_all_uppers*dt_reds, reds_t2_all_lowers*dt_reds, color=colors[3], alpha=0.2)

plt.xlabel('Lagtime (ns)')
plt.ylabel('Implied timescale (ns)')
plt.yscale('log')
plt.ylim(1, 1e5)
plt.legend()
plt.title('Implied Timescales Comparison')
plt.grid(True)
plt.tight_layout()
plt.show()
```
