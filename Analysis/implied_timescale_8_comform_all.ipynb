## Load Important Functoin

```python
class Data:
    def __init__(self, array_list):
        self.array_list = array_list

    def save(self, filename):
        with open(filename, 'wb') as f:
            pickle.dump(self.array_list, f)

    @classmethod
    def load(cls, filename):
        with open(filename, 'rb') as f:
            array_list = pickle.load(f)
        return cls(array_list)
    
```

### omg8rep

```python
from tqdm import tqdm

methods_path = f"{metadynamics_path}/autocooreldata/autocoorel_data_omg8rep"

meta_omg8rep_t2_means = []
meta_omg8rep_t2_stds =[]

for i in tqdm(range(8)):
    omega_loaded_single = Data.load(f"{methods_path}/stationary_distributions{i}.pkl").array_list
    t2_means_single = omega_loaded_single[4]
    t2_uppers_single = omega_loaded_single[5]
    t2_lowers_single = omega_loaded_single[6]
    t2_stds_single = omega_loaded_single[7]

    meta_omg8rep_t2_means.append(t2_means_single)
    meta_omg8rep_t2_stds.append(t2_stds_single)

meta_omg8rep_lagtimes_ns = omega_loaded_single[8]

#meta_omg8rep_pi_all_means = meta_omg8rep_pi_means
meta_omg8rep_t2_means = np.array(meta_omg8rep_t2_means)
meta_omg8rep_t2_stds = np.array(meta_omg8rep_t2_stds)


# Mean across all replicas
meta_omg8rep_t2_all_means = np.nanmean(meta_omg8rep_t2_means, axis=0)

# Law of total variance
within_var = np.nanmean(meta_omg8rep_t2_stds**2, axis=0)  # average within-replica variance
between_var = np.nanvar(meta_omg8rep_t2_means, axis=0, ddof=1)  # var
meta_omg8rep_t2_all_stds = np.sqrt(within_var + between_var)

meta_omg8rep_lagtimes_ns[-1]
```

```python
dt_in_ns = 0.0004

plt.figure()
plt.plot(meta_omg8rep_lagtimes_ns, meta_omg8rep_t2_all_means*dt_in_ns, 'ko-', alpha=1)
plt.fill_between(meta_omg8rep_lagtimes_ns, 
                (meta_omg8rep_t2_all_means + meta_omg8rep_t2_all_stds) *dt_in_ns, 
                 (meta_omg8rep_t2_all_means - meta_omg8rep_t2_all_stds) *dt_in_ns, alpha=0.3)
plt.xlabel('lagtime (ns)')
plt.ylabel('implied timescale (ns)')

plt.yscale('log')
plt.ylim(0.001, 100)
plt.title(f'Implied timescales for Plain MD')
plt.show()
```
